*Los sistemas de control de versiones* (VCS) son herramientas que se utilizan para realizar un seguimiento de los cambios en el cdigo fuente y facilitar la colaboracin. Mantienen un historial de cambios en una serie de instantneas, que son encapsulaciones del estado completo de archivos/carpetas dentro de un directorio de nivel superior, junto con metadatos como el autor del cambio, mensajes asociados con el cambio y ms. Git es el VCS ms popular y su modelo de datos consta de instantneas en forma de *grficos acclicos dirigidos* (DAG) de confirmaciones, donde cada confirmacin se refiere a un conjunto de instantneas principales. Las confirmaciones en Git son inmutables, pero los errores an se pueden corregir creando nuevas confirmaciones. El diseo de Git se basa en un modelo de datos bien pensado, que habilita funciones como mantener el historial, admitir sucursales y permitir la colaboracin.
