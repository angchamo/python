*Los sistemas de control de versiones* (VCS) son herramientas que se utilizan para realizar un seguimiento de los cambios en el código fuente y facilitar la colaboración. Mantienen un historial de cambios en una serie de instantáneas, que son encapsulaciones del estado completo de archivos/carpetas dentro de un directorio de nivel superior, junto con metadatos como el autor del cambio, mensajes asociados con el cambio y más. Git es el VCS más popular y su modelo de datos consta de instantáneas en forma de *gráficos acíclicos dirigidos* (DAG) de confirmaciones, donde cada confirmación se refiere a un conjunto de instantáneas principales. Las confirmaciones en Git son inmutables, pero los errores aún se pueden corregir creando nuevas confirmaciones. El diseño de Git se basa en un modelo de datos bien pensado, que habilita funciones como mantener el historial, admitir sucursales y permitir la colaboración.
