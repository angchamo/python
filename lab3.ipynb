1) ¿Cuál es el tamaño en bytes de un array de 1000  
elementos de tipo booleano?

4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], 
obtenga la media aritmética de la diferencia (a-b)

```python
# Crear los arrays
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calcular la diferencia entre los arrays
diff = a - b

# Calcular la media aritmética de la diferencia
mean_diff = np.mean(diff)

print(f"La media aritmética de la diferencia (a-b) es: {mean_diff}")
```

5) Tengo valores de cordenadas (x, y) en las columnas 
del array

30.0 10.7 

70.2 0.

Agregue a este array las coordenadas (37.1, -3.6). 
Muestre en pantalla las dimensiones del nuevo 
array.

```python
# Crear el array de coordenadas existente
coordenadas = np.array([[1.33, 4.5],
                        [30.0, 10.7],
                        [70.2, 0.5]])

# Agregar las nuevas coordenadas
nuevas_coordenadas = np.array([[37.1, -3.6]])
coordenadas = np.vstack((coordenadas, nuevas_coordenadas))

# Mostrar las dimensiones del nuevo array
print("El nuevo array tiene las siguientes dimensiones:")
print(coordenadas.shape)
```

5) Obtenga la media, mediana, desviación estándar, valor 
máximo y valor mínimo de la población de los municipios 
de la provincia de Granada. 

```python
granada = df[df['Provincia'] == 'Granada']
poblacion = granada['Poblacion']

med = np.median(poblacion)
dev = np.std(poblacion)
max = np.max(poblacion)
min = np.min(poblacion)

print(f'mediana: {med}, desviacion estandar: {dev}, maximo: {max}, minimo: {min} ')
```
